```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv('Output_data.csv')
data.head()
```

```python
data['pass'].value_counts().plot(kind = 'bar')
plt.xlabel('Pass(1) / Fail(0)')
plt.ylabel('Number Of Student')
plt.title('Student Pass-Fail Table- Count')
plt.show()
```

```python
sns.boxplot(x='pass',y='attendance_pct',data = data)
plt.title('Attendence VS Pass/Fail')
plt.show
```

```python
sns.boxplot(x='pass', y ='study_hours_per_week',data =data)
plt.title('Study Hours BS VS Pass Fail')
```

```python
sns.scatterplot(y='study_hours_per_week', x='midterm_score',data= data,hue = 'pass')
plt.title('Study Hours VS Midterm Score')
plt.xlabel('Midterm Score')
plt.ylabel('Study Hours')
plt.show
```

```python
# Correlation Heat map

corr = data.corr()

sns.heatmap(corr, annot = True, cmap ='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
